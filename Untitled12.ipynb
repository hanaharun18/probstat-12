```python
# Hana Zalfa Azizah - 064102400007
import numpy as np
import matplotlib.pyplot as plt

# Data
ukuran = np.array([40, 55, 60, 70, 85, 100, 120, 150, 200, 250])
harga = np.array([3.0, 4.2, 4.5, 5.2, 6.0, 7.2, 8.4, 10.5, 14.2, 17.0])

# Scatter plot
plt.scatter(ukuran, harga, color='darkgreen')
plt.title("Ukuran Apartemen vs Harga Sewa")
plt.xlabel("Ukuran (mÂ²)")
plt.ylabel("Harga Sewa (juta/bulan)")
plt.grid(True)
plt.show()
```

```python
# Hana Zalfa Azizah - 064102400007
from sklearn.linear_model import LinearRegression

# Ubah ukuran jadi matriks kolom
X = ukuran.reshape(-1, 1)

# Buat model regresi linier
model = LinearRegression()
model.fit(X, harga)

# Prediksi harga dengan model regresi
prediksi_harga = model.predict(X)
```

```python
# Hana Zalfa Azizah - 064102400007
import matplotlib.pyplot as plt

plt.scatter(ukuran, harga, label="Data Asli", color='darkgreen')
plt.plot(ukuran, prediksi_harga, color='orange', label="Garis Regresi")
plt.xlabel("Ukuran (mÂ²)")
plt.ylabel("Harga Sewa (juta/bulan)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Hana Zalfa Azizah - 064102400007
import numpy as np

corr = np.corrcoef(ukuran, harga)[0, 1]
print(f"Korelasi Pearson: {corr:.3f}")
```

```python
# Hana Zalfa Azizah - 064102400007
from sklearn.metrics import r2_score

r2 = r2_score(harga, prediksi_harga)
print(f"Koefisien Determinasi (RÂ²): {r2:.3f}")
```

```python
# Hana Zalfa Azizah - 064102400007
x_baru = np.array([[175]])  # Ukuran baru dalam format 2D
y_baru = model.predict(x_baru)
print(f"Prediksi harga sewa untuk 175 mÂ²: {y_baru[0]:.2f} juta/bulan")
```
